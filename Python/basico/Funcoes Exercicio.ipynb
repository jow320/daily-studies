# Exercicios Funções

```python
#Função que calcule o quadrado de um numero, tendo como parametro padrão o número 5

def quadrado(numero = 5):
  "Função que calcula o quadrado de um número"
  return(numero * numero)

quadrado(4)
```

```python
#Função que converte Fahrenheit para Celsius

def conversorFtoC(temp):
  "Função que converte Fahrenheit para Celsius"
  resultado = (temp - 32) * 5/9
  return (resultado)

conversorFtoC(100)

```

```python
#Função que calcula Quadrado ou cubo do segundo parametro.

def quadradoOuCubo(palavra, number):
  "Função que calcula Quadrado ou cubo do segundo parametro"
  if palavra == "quadrado":
    return(number * number)
  elif palavra == "cubo":
    return(number * number * number)
  else:
    return(print("Insira parametros válidos como ex: (cubo, 5) ou (quadrado, 4)"))

quadradoOuCubo("rasteira", 5)
```
