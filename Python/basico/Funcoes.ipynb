# Funções

```python
#função sem retorno e sem passagem de parametro
def ola_mundo():
  "Função que mostra uma mensagem padrão na tela"
  print("Olá mundo!")

ola_mundo()
help(ola_mundo)
```

```python
#função sem retorno e com passagem de parametro
def ola_mundo(name):
  "Função que mostra uma mensagem customizada na tela"
  print(f"Olá mundo! meu nome é {name}!")

nome = input("Qual seu nome?")
ola_mundo(nome)
```

```python
#Função com retorno e com passagem de parametro
def soma(a,b):
  "Função que realiza uma soma entre dois numeros inseridos"
  return (a+b)

resultado = soma(5,5)
resultado
```
