```python
#Fatorial de um número, ex: 5! = 5*4*3*2*1
```

```python
numero1 = eval(input("Qual o número?"))
numero2 = numero1
while numero1 > 0:
  print(numero2, end = " ")
  numero1 -= 1
  numero2 =  numero2 * numero1
```
