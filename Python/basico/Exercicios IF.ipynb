# Exercícios IF

```python
nota1 = eval(input('Qual a primeira nota?'))
nota2 = eval(input("Qual a segunda nota?"))
media = (nota1 + nota2) / 2
if media > 7:
  print(f"Você foi aprovado com média {media}!")
else:
  print(f"Você foi reprovado com média {media}!")
```

```python
nota1 = eval(input('Qual a primeira nota?'))
nota2 = eval(input("Qual a segunda nota?"))
media = (nota1 + nota2) / 2
if media >= 7:
  print(f"Você foi aprovado com média {media}, Parabéns!")
elif media >= 5 and media < 7:
  print(f"Você está de exame com média {media}")
else:
  print(f"Você foi reprovado com média {media}")
```
