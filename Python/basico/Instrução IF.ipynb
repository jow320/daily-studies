# Controle de Decisão - Instrução IF

```python
if <condição>:
  <bloco de códigos>
<instruções>
```

```python
if <condição>:
  <bloco de códigos>
elif <condição2>:
  <bloco de códigos>
elif <condição2>:
  <bloco de códigos>
else:
  <ultimo bloco>
<instruções>
```

```python
temp = eval(input("Qual a temperatura?")) #Dizer se está frio ou não.
if temp >= 40:
  print("Está muito quente")
elif temp > 30:
  print("Está agradável")
else:
  print("Está frio")
```
