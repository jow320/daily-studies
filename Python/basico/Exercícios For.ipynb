```python
#tabuada
numero1 = int(input("Qual o número?"))
numero2 = int(input("Até onde vai a tabuada?"))

for tabuada in range(0, numero2 + 1):
  print(f"{numero1} * {tabuada} = {numero1 * tabuada}")
```

```python
#elemento fibonacci
n = int(input("Qual o elemento a ser mostrado?"))

primeiro = 1
segundo = 1

if n == 1 or n == 2:
  print(1)
else:

 for _ in range(2, n):
  elemento = primeiro + segundo
  segundo = primeiro
  primeiro = elemento
  print(elemento)

```
