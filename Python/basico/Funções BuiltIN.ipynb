# Funções Built In

```python
abs(-10) # Retorna o valor absoluto
```

```python
precoMinimo = min(10, 20, 30, 40) # Retorna o valor minimo de uma sequência
precoMinimo
```

```python
precoMax = max(10, 20, 30, 40) # Retorna o valor maximo de uma sequência
precoMax
```

```python
somaTotal = sum([10, 20, 30, 40]) # Retorna a soma de uma sequência
somaTotal
```

```python
exponencial = pow(3, 3) # Retorna um exponente
exponencial
```
