# Laço while

```python
while condicao:
  comandos1
  comandos2
```

```python
numero = 0
while numero <= 10:
  print(numero)
  numero += 1 #numero = numero + 1
print("Acabou o loop while")
```

```python
numero_max = eval(input("Qual o número máximo?"))
numero = 0
while numero <= numero_max:
  print(numero)
  numero += 1

```

# Laço For

```python
for variavel in itens:
  comando1
  comando2
  comando3
```

```python
for dia in ["Segunda", "Terça", "Quarta", "Quinta", "Sexta"]:
  print(dia)
```

```python
for numero in [1 , 2 , 3, 4, 5, 6]:
  print(numero,end=" ")
```

```python
for numero in range(6): #1 ao 5, em listas o 0 conta como um número, simbolizando o 1 da sequência.
  print(numero,end=" ")
```

```python
for numero in range(5,20): #5 ao 19
  print( numero, end=" ")
```

```python
for numero in range(5, 20, 5): #5 ao 19 de 5 em 5
  print( numero, end=" ")
```
